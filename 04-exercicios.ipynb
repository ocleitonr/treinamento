### 1. Some todos os números de 1 até 100

```python
i = 1
soma = 0

for i in range(100):
  soma += i
print(soma)
```

### 2. Some todos os números pares de 1 até 100

```python
soma = 0
impar = 1
for i in range(100):
     soma += impar
     impar += 1
     print(soma)
```

### 3. Some todos os números ímpares de 1 até 100

```python
soma = 0
impar = 1
for i in range(100):
     soma += impar
     impar += 1
     print(soma)
```

### 4. Dada a lista abaixo, some todos os elementos inteiros positivos

```python
lista = [-2, 2, 'celular', 10, -99, False]
lista
```

```python
for lista in range(lista):
  if lista >= 1:
    soma = lista + lista
    print(soma)
```
